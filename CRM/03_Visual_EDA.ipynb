```python
# ============================================================
# SalesFlow Analytics
# Visual EDA Dashboard
# Korean Startup Style CRM Analytics
# ============================================================

# =========================
# IMPORT LIBRARIES
# =========================

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

# =========================
# STYLE SETTINGS
# =========================

plt.style.use("default")

pd.set_option("display.max_columns", None)

# =========================
# LOAD CLEAN DATASET
# =========================

print("=" * 60)
print("Loading Clean Dataset")
print("=" * 60)

df = pd.read_excel(
    r"C:\Users\Дима\Desktop\main imir\Python\CRM_new\AdventureWorks Sales Sample.xlsx",
    sheet_name="Sales_data"
)

print("\nDataset Loaded Successfully")
```

```python
# =========================
# DASHBOARD INTRO
# =========================

print("\n" + "=" * 60)
print("Customer Intelligence Dashboard")
print("=" * 60)

print("""
This dashboard explores:

- customer behavior
- regional demand
- product trends
- sales patterns
- business insights
""")
```

```python
# ============================================================
# SECTION 1
# KPI OVERVIEW
# ============================================================

total_sales = df["Sales Amount"].sum()

total_orders = len(df)

average_sales = df["Sales Amount"].mean()

unique_products = df["ProductKey"].nunique()

print("\n" + "=" * 60)
print("KPI Overview")
print("=" * 60)

print(f"\nTotal Revenue      : ${total_sales:,.2f}")
print(f"Total Orders       : {total_orders:,}")
print(f"Average Sale       : ${average_sales:,.2f}")
print(f"Unique Products    : {unique_products:,}")
```

```python
# =========================
# KPI VISUALIZATION
# =========================

kpi_names = [
    "Revenue",
    "Orders",
    "Avg Sales",
    "Products"
]

kpi_values = [
    total_sales / 1000000,
    total_orders / 1000,
    average_sales,
    unique_products
]

plt.figure(figsize=(12, 6))

bars = plt.bar(
    kpi_names,
    kpi_values,
    color=[
        "#DDD6FE",
        "#C4B5FD",
        "#A78BFA",
        "#8B5CF6"
    ],
    width=0.55
)

plt.title(
    "Business KPI Overview",
    fontsize=22,
    fontweight="bold",
    color="#1F2937",
    pad=20
)

plt.grid(
    axis="y",
    linestyle="--",
    alpha=0.3
)

for bar in bars:
    height = bar.get_height()

    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        f"{height:.1f}",
        ha="center",
        va="bottom",
        fontsize=11
    )

plt.show()
```

```python
# ============================================================
# SECTION 2
# TERRITORY SALES ANALYSIS
# ============================================================

print("\n" + "=" * 60)
print("Territory Sales Analysis")
print("=" * 60)

territory_sales = df.groupby(
    "SalesTerritoryKey"
)["Sales Amount"].sum().sort_values()
```

```python
# =========================
# TERRITORY SALES CHART
# =========================

plt.figure(figsize=(12, 6))

bars = plt.barh(
    territory_sales.index.astype(str),
    territory_sales.values,
    color="#A78BFA"
)

plt.title(
    "Sales Distribution by Territory",
    fontsize=20,
    fontweight="bold",
    color="#1F2937",
    pad=20
)

plt.xlabel("Sales Amount")

plt.grid(
    axis="x",
    linestyle="--",
    alpha=0.3
)

plt.show()
```

```python
print("""
Insight:
Some sales territories generate significantly higher revenue,
indicating stronger regional purchasing activity.
""")
```

```python
# ============================================================
# SECTION 3
# PRODUCT TRENDS
# ============================================================

print("\n" + "=" * 60)
print("Product Trend Analysis")
print("=" * 60)

top_products = df.groupby(
    "ProductKey"
)["Sales Amount"].sum().sort_values(
    ascending=False
).head(10)
```

```python
# =========================
# TOP PRODUCTS GRAPH
# =========================

plt.figure(figsize=(12, 6))

bars = plt.bar(
    top_products.index.astype(str),
    top_products.values,
    color="#8B5CF6"
)

plt.title(
    "Top Performing Products",
    fontsize=20,
    fontweight="bold",
    color="#1F2937",
    pad=20
)

plt.xticks(
    rotation=45,
    ha="right"
)

plt.grid(
    axis="y",
    linestyle="--",
    alpha=0.3
)

plt.show()
```

```python
print("""
Insight:
Several products generate significantly higher revenue,
indicating concentrated product demand behavior.
""")
```

```python
# ============================================================
# SECTION 4
# COST VS REVENUE ANALYSIS
# ============================================================

print("\n" + "=" * 60)
print("Cost vs Revenue Analysis")
print("=" * 60)

plt.figure(figsize=(12, 6))

plt.scatter(
    df["Total Product Cost"],
    df["Sales Amount"],
    alpha=0.4,
    color="#8B5CF6"
)

plt.title(
    "Cost vs Sales Revenue",
    fontsize=20,
    fontweight="bold",
    color="#1F2937",
    pad=20
)

plt.xlabel("Total Product Cost")
plt.ylabel("Sales Amount")

plt.grid(
    linestyle="--",
    alpha=0.3
)

plt.show()
```

```python
print("""
Insight:
Higher product costs are generally associated with
higher sales revenue, indicating premium product behavior.
""")
```

```python
# ============================================================
# SECTION 5
# ORDER QUANTITY DISTRIBUTION
# ============================================================

print("\n" + "=" * 60)
print("Order Quantity Distribution")
print("=" * 60)

plt.figure(figsize=(12, 6))

plt.hist(
    df["Order Quantity"],
    bins=30,
    color="#A78BFA",
    alpha=0.8
)

plt.title(
    "Order Quantity Distribution",
    fontsize=20,
    fontweight="bold",
    color="#1F2937",
    pad=20
)

plt.xlabel("Order Quantity")
plt.ylabel("Frequency")

plt.grid(
    linestyle="--",
    alpha=0.3
)

plt.show()
```

```python
print("""
Insight:
Most customer orders are concentrated within lower
quantity ranges, indicating frequent small purchases.
""")
```

```python
# ============================================================
# SECTION 6
# CUSTOMER PURCHASE ACTIVITY
# ============================================================

print("\n" + "=" * 60)
print("Customer Purchase Activity")
print("=" * 60)

customer_sales = df.groupby(
    "CustomerKey"
)["Sales Amount"].sum().sort_values(
    ascending=False
).head(15)
```
